```python
means= df.loc[:, '1990':'2020'].mean()
stds = df.loc[:, '1990':'2020'].std()
vars = df.loc[:, '1990':'2020'].var()
```

```python
data_description = pd.DataFrame(data={
                    'mean':means,
                    'std_deviation':stds,
                    'varience':vars},index=means.index)
```

```python
df.head()
```

```python
mean_emissions = means
std_emissions = stds

plt.plot(mean_emissions.index, mean_emissions.values, marker='o', linestyle='-', color='b', label='Mean')

plt.fill_between(mean_emissions.index, mean_emissions - 3 * std_emissions, mean_emissions + 3 * std_emissions, color='b', alpha=0.2, label='Â±3 Std Dev')

plt.title('Mean CO2 Emissions with 3x Standard Deviation Over the Years')
plt.xlabel('Year')
plt.ylabel('Mean CO2 Emission')
plt.xticks(rotation='vertical')  # Display years vertically
plt.legend()
plt.grid(True)
plt.show()
```
